```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
def norm(s):
    return (s - s.min()) / (s.max() - s.min())
```

```python
xas_exp = pd.read_csv('XAS_Exp.dat', sep='\s+', header=None, names=['Energy', 'Intensity'])
xas_fit = pd.read_csv('XAS_Fit.dat', sep='\s+', header=None, names=['Energy', 'Intensity'])

```

```python
fig, ax = plt.subplots()
ax.plot(xas_exp['Energy'], xas_exp['Intensity'], label='Experiment')
ax.plot(xas_fit['Energy'], xas_fit['Intensity'], label='Fit')
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('Intensity (a.u.)')
ax.legend()
```

```python
rixs_exp = pd.read_csv('RIXS_Exp_Trimmed.dat', sep='\s+', header=None, names=['Energy', 'Intensity1', 'Intensity2', 'Intensity3', 'Intensity4'])
rixs_fit = pd.read_csv('RIXS_Fit.dat', sep='\s+', header=None, names=['Energy', 'Intensity1', 'Intensity2', 'Intensity3', 'Intensity4'])
```

```python
emis_dict = {1: '638', 2: '639.35', 3: '640.16', 4: '640.75'}
```

```python
rixs_comp_fig, rixs_comp_ax = plt.subplots()
for i in range(1,5):
    rixs_comp_ax.set_prop_cycle(None)
    rixs_comp_ax.plot(rixs_exp['Energy'], norm(rixs_exp[f'Intensity{i}'])+i-1)
    rixs_comp_ax.text(0.8, 1.05 - 0.22*i, f'{emis_dict[5-i]} eV', ha='left', va='top', size=12, transform=rixs_comp_ax.transAxes)
    rixs_comp_ax.plot(rixs_fit['Energy'], norm(rixs_fit[f'Intensity{i}'])+i-1)

rixs_comp_ax.set_xlim(-5,20)
rixs_comp_ax.set_xlabel('Energy (eV)')
rixs_comp_ax.set_ylabel('Intensity (a.u.)')


custom_lines = [Line2D([0], [0], color='#1f77b4', lw=1.5),
                Line2D([0], [0], color='#ff7f0e', lw=1.5),
                Line2D([0], [0], color='#2ca02c', lw=1.5),
                Line2D([0], [0], color='#d62728', lw=1.5),
                Line2D([0], [0], color='#9467bd', lw=1.5)]
# rixs_comp_ax.legend(custom_lines, ['Exp', 'Mn_Oh_2', 'Mn_Oh_3', 'Mn_Td_2', 'Mn_Td_3'])
```
